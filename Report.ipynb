## Validation curve

So maybe our bad results are due to a hyperparameter in the classifier. Let's see if we can find the optimal value for the regularization parameter $\alpha$:

```python
from sklearn.learning_curve import validation_curve

def plot_validation_curve(estimator, title, X, y, param_name, param_range,
                          ylim=None, cv=None, n_jobs=1):

    train_scores, test_scores = validation_curve(
    estimator, X, y, param_name=param_name, param_range=param_range,
    cv=cv, scoring="f1_weighted", n_jobs=n_jobs)

    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.title(title)
    plt.xlabel(param_name)
    plt.ylabel("F1 score")
    if ylim != None:
        plt.ylim(ylim)
    plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
    plt.fill_between(param_range, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.2, color="r")
    plt.semilogx(param_range, test_scores_mean, label="Cross-validation score",
                 color="g")
    plt.fill_between(param_range, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.2, color="g")
    plt.legend(loc="best")
    
    return plt
```

```python
%%time
param_name = "alpha"
param_range = np.logspace(-49, 5, 11)
title = "Validation Curve with SGD"
plot_validation_curve(SGDclfPL, title, X, y, param_name, param_range,
                      cv=cv, n_jobs=1)
plt.show()
```

It is definitely possible to get a **worse** score by tuning the regularization parameter, but the default value of $\alpha = 10^{-4}$ seems to be as good as it gets.

But it is a very strange looking validation curve. I would like it to look something like this:

In the validation curve above, we see that for low $\gamma$ values both the training and CV score is low because the model is **too constrained** to fit the data properly. For high $\gamma$ values, the model is **too loose** so that the model overfits the training data and is not able to *generalize* to the test data, so the test score plummets.

But the good-looking validation curve was made with SVM and not SGD, so I ran the same code that generated the good-looking graph with SGD instead of SVM and got this graph:

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.learning_curve import validation_curve

digits = load_digits()
X, y = digits.data, digits.target

param_range = np.logspace(-49, 5, 11)
train_scores, test_scores = validation_curve(
    SGDClassifier(), X, y, param_name="alpha", param_range=param_range,
    cv=10, scoring="f1_weighted", n_jobs=1)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

plt.title("Validation Curve with SGD")
plt.xlabel("alpha")
plt.ylabel("Score")
plt.ylim(0.0, 1.1)
plt.semilogx(param_range, train_scores_mean, label="Training score", color="r")
plt.fill_between(param_range, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.2, color="r")
plt.semilogx(param_range, test_scores_mean, label="Cross-validation score",
             color="g")
plt.fill_between(param_range, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.2, color="g")
plt.legend(loc="best")
plt.show()
```

That looks like the validation curve for my data! (Except the score is much better...)
So maybe we should not be so upset about the strange-looking validation curve. We turn our attention to the learning curve instead.

## Learning Curve

```python
# See examples/model_selection/plot_learning_curve.py

from sklearn import cross_validation
from sklearn.learning_curve import learning_curve


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.01, 1.0, 20)):

    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("F1 score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, scoring="f1_weighted",
        n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
%%time
title = "Learning Curves (SGD)"
plot_learning_curve(SGDclfPL, title, X, y, cv=cv, n_jobs=5)
plt.show()
```

At the first data point, where we use only 1% of our training examples, we see that the Training score outperforms the CV score (although they are both very bad). As we go towards using 100% of our data, the two scores align. That means that we will not benefit by getting more training examples.

This is bad.

Maybe it's because our data is not linearly separable, so the more examples we look at, the more wrong we are.

## Non-linear data and kernel approximation

Let's create some non-linearly separable data:

```python
from sklearn.datasets import make_blobs
centers = np.array([[10, 10], [-10, -10], [10, -10], [-10, 10]])
X_blobs, y_blobs = make_blobs(n_samples=400, n_features=2,
                              centers=centers, cluster_std=4.0)
y_blobs[y_blobs==0] = 1
y_blobs[y_blobs==1] = 1
y_blobs[y_blobs==2] = -1
y_blobs[y_blobs==3] = -1
```

```python
plt.scatter(X_blobs[:,0], X_blobs[:,1], c=y_blobs, cmap=plt.cm.Paired)
plt.show()
```

First try to use a linear SVM:

```python
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
lin_svc = LinearSVC()
scores = cross_val_score(lin_svc, X_blobs, y_blobs, cv=5,
                         scoring="f1_weighted")
print "F1: %.2f ± %.2f" % (scores.mean(), scores.std())
```

As expected, the score is around 50%. There just is no way to separate these points with a straight line.

But an RBF kernel, which is a non-linear SVM should be able to do it.

```python
from sklearn.svm import SVC
from sklearn.cross_validation import cross_val_score
rbf_svc = SVC(kernel="rbf")
scores = cross_val_score(rbf_svc, X_blobs, y_blobs, cv=5,
                         scoring="f1_weighted")
print "F1: %.2f ± %.2f" % (scores.mean(), scores.std())
```

Yeah, it can almost perfectly separate the points without any tweaking whatsoever.

But as my data set is so large, using an RBF kernel would take a good part of forever.

Instead, let's try kernel approximation with the Nystroem method!

```python
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
lin_svc = LinearSVC()
feature_map_nystroem = Nystroem()
nystroem_approx_svm = Pipeline([("feature_map", feature_map_nystroem),
                                        ("svm", lin_svc)])
scores = cross_val_score(nystroem_approx_svm, X_blobs, y_blobs, cv=5,
                         n_jobs=5, scoring="f1_weighted")
print "F1: %.2f ± %.2f" % (scores.mean(), scores.std())
```

With that feature mapping trick we got a very good score using a **linear** SVM.

Let's try that with our own data!

```python
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline

lin_svc = LinearSVC()
feature_map_nystroem = Nystroem()
nystroem_approx_svm = Pipeline([("feature_map", feature_map_nystroem),
                                        ("svm", lin_svc)])
scores = cross_val_score(nystroem_approx_svm, X, y, cv=cv,
                         n_jobs=5, scoring="f1_weighted")
print "F1: %.2f ± %.2f" % (scores.mean(), scores.std())
```
